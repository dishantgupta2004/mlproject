```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from ucimlrepo import fetch_ucirepo 
```

```python
# fetch dataset 
energy_efficiency = fetch_ucirepo(id=242) 
# data (as pandas dataframes) 
column_X = ['Relative_Compactness', 'Surface_Area', 'Wall_Area', 'Roof_Area', 'Overall_Height', 'Orientation', 'Glazing_Area', 'Glazing_Area_Distribution']
column_Y = ['Heating_Load', 'Cooling_Load']
X = energy_efficiency.data.features 
y = energy_efficiency.data.targets 
df_X = pd.DataFrame(data= X)
df_y = pd.DataFrame(data= y )
df_X.columns = column_X
df_y.columns = column_Y
```

```python
for col in df_X.columns:
    plt.figure(figsize= (8, 6))
    sns.displot(df_X[col], kde= True)
    plt.xlabel(col)
    plt.ylabel("Target_variables")
    plt.legend()
    plt.show()
```

```python
df_X.describe()

## Almost constant values in above graphs  (more frequent common)
```

```python
### Outlier detection
for col in df_X.columns:
    sns.boxplot(x= df_X[col], y= df_y["Cooling_Load"])
    plt.xlabel(col)
    plt.ylabel("Cooling Load")
    plt.show()
```

```python
### Outlier detection
for col in df_X.columns:
    sns.boxplot(x= df_X[col], y= df_y["Heating_Load"])
    plt.xlabel(col)
    plt.ylabel("Heating Load")
    plt.show()
```

```python
## No outliers only some insignificant outliers present that will does not impact our prediction
### Covariance checking

df_X.cov()
```

```python
### Variance Inflation Factor (for multicollinearity)

from statsmodels.stats.outliers_influence import variance_inflation_factor
vif_data = pd.DataFrame()
vif_data["feature"] = df_X.columns
vif_data["VIF"] = [variance_inflation_factor(df_X.values, i) for i in range(len(df_X.columns))]
```
